```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.quiver(0,0,4,4)  ## Use to plot vector
```

```python
plt.quiver(0,0,4,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
```

```python
def plot_vectors(vects, color=None):
    for i,vec in enumerate(vects):
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units='xy', angles='xy',scale=1,color=color[i])
    plt.xlim(-10,10)
    plt.ylim(-10,10)
```

```python
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,3,-4)],['r','g','y'])
```
