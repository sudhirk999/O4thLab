```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_vectors(vects):
    colors = ['r','b','g','y']
    for i,vec in enumerate(vects):
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units='xy', angles='xy',scale=1,color=colors[i])
    plt.xlim(-10,10)
    plt.ylim(-10,10)
```

```python
vecs = [np.array([0,0,3,4]), np.array([0,0,-3,4]),np.array([0,0,-3,5]),np.array([0,0,-5,-4])]
```

```python
plot_vectors([vecs[0], vecs[2]])
```

```python
plot_vectors([vecs[0], vecs[2], vecs[0] + vecs[2]])
```

```python
plot_vectors([vecs[0], vecs[3], vecs[0] + vecs[3]])
```

```python
## SUBTRACTION

plot_vectors([vecs[0], vecs[2], vecs[0] - vecs[2]])
```

```python
plot_vectors([vecs[0], -vecs[2], vecs[0] + (-vecs[2])])
```
